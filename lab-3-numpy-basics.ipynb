# Measure of Central Tendency

```python
resistance = np.array([
    [10.6, 9.1, 9.3, 9.8, 10.5, 10.4, 9.5, 11, 10.4, 3, 9.8], # Jaguar
    [10.1, 11, 9.1, 20, 9.2, 10.8, 9.9, 9.2, 9.1, 9.1, np.nan]  # Panther
])
resistance
```

```python
np.shape(resistance)
```

```python
resistance.T
```

```python
np.mean(resistance, axis=0)
```

```python
np.mean(resistance, axis=1)
```

```python
#Mean
np.nanmean(resistance, axis = 1)
```

```python
jag_mean = np.nanmean(resistance, axis=0)[0]
pan_mean = np.nanmean(resistance, axis=1)[1]

print(f'Jaguar_Mean = {jag_mean:.2f} ohms')
print(f'Panther_Mean = {pan_mean:.2f} ohms')
```

```python
#Meadian
np.nanmedian(resistance, axis=1)
```

```python
jag_median = np.nanmedian(resistance, axis=0)[0]
pan_median = np.nanmedian(resistance, axis=1)[1]

print(f'Jaguar_Median = {jag_median:.2f} ohms')
print(f'Panther_Median = {pan_median:.2f} ohms')
```

```python
#Mode
stats.mode(resistance, axis=1)
```

```python
my_array = [1,1,2,2,3,4,5]
multimode(my_array)
```

```python
jag_res = resistance[0]

print("Jaguar_Mode:")
for i in multimode(jag_res):
    print(f'{i} ohms')
```

```python
pan_res = resistance[1]

print("Panther_Mode:")
for i in multimode(pan_res):
    print(f'{i} ohms')
```

# Measures of Variability

```python
#Jaguar vs Panther
resistance = np.array([
    [10.6, 9.1, 9.3, 9.8, 10.5, 10.4, 9.5, 11, 10.4, 3, 9.8], # Jaguar
    [10.1, 11, 9.1, 20, 9.2, 10.8, 9.9, 9.2, 9.1, 9.1, np.nan]  # Panther
])
resistance
```

```python
#Range
res_max = np.nanmax(resistance, axis=1)
res_min = np.nanmin(resistance, axis=1)

res_range = res_max - res_min
res_range
```

```python
print(f'Jaguar_Range = {res_range[0]:.2f} ohms')
print(f'Panther_Range = {res_range[1]:.2f} ohms')
```

```python
#1st Quartile
res_Q1 = np.nanpercentile(resistance,25,axis=1)
res_Q1
```

```python
#2nd Quartile
res_Q2 = np.nanpercentile(resistance,50,axis=1)
res_Q2
```

```python
#3rd Quartile
res_Q3 = np.nanpercentile(resistance,75,axis=1)
res_Q3
```

```python
#IQR
print(f'Jaguar_IQR = {res_Q3[0] - res_Q1[0]:.2f} ohms')
print(f'Panther_IQR = {res_Q3[1] - res_Q1[1]:.2f} ohms')
```

```python
#Variance
res_var = np.nanvar(resistance,ddof=1,axis=1)
res_var
```

```python
print(f'Jaguar_Variance = {res_var[0]:.2f} ohms')
print(f'Panther_Variance = {res_var[1]:.2f} ohms')
```

```python
#Standar Deviation
res_std = np.nanstd(resistance,ddof=1,axis=1)
res_std
```

```python
print(f'Jaguar_Standard Deviation = {res_std[0]:.2f} ohms')
print(f'Panther_Standard Deviation = {res_std[1]:.2f} ohms')
```
