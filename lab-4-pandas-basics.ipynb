```python
#imports and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# DataFrame

```python
#Creating a Dictionary
data = {
    "Name":['Henry','Owen','Ada'],
    "Age":[22,35,58],
    "Sex":['M','M','F']
}
```

```python
#Creating DataFrame from Dictionary
df = pd.DataFrame(data)
df
```

# Descriptive Statistics

```python
#Jaguar vs Panther
resistance = np.array([
    [10.6, 9.1, 9.3, 9.8, 10.5, 10.4, 9.5, 11, 10.4, 3, 9.8], # Jaguar
    [10.1, 11, 9.1, 20, 9.2, 10.8, 9.9, 9.2, 9.1, 9.1, np.nan]  # Panther
]).T
resistance
```

```python
#Creating a DataFrame from numpy array
df = pd.DataFrame(resistance,columns=['Jaguar', 'Panther'])
df
```

```python
#Summary of DataFrame
df.info()
```

```python
type(df)
```

```python
#Summary of Statistics
df.describe()
```

```python
#Median
jag_median = df['Jaguar'].median()
pan_median = df['Panther'].median()

print(f'Jaguar_Median = {jag_median:.2f} ohms')
print(f'Panther_Median = {pan_median:.2f} ohms')
```

```python
#Mode
jag_mode = df['Jaguar'].mode()
pan_mode = df['Panther'].mode()
```

```python
jag_mode
```

```python
jag_mode.info()
```

```python
type(jag_mode)
```

```python
print('Jaguar_Mode:')

for i in jag_mode:
    print(i)
```

```python
print('Panther_Mode:')

for i in pan_mode:
    print(i)
```

```python
#Jaguar IQR
Q1 = df['Jaguar'].quantile(0.25)
Q3 = df['Jaguar'].quantile(0.75)

print(f'Jaguar_IQR = {Q3-Q1:.2f} ohms')
```

```python
#Panther IQR
Q1 = df['Panther'].quantile(0.25)
Q3 = df['Panther'].quantile(0.75)

print(f'Panther_IQR = {Q3-Q1:.2f} ohms')
```

# Bar Chart

```python
jag_mean = df['Jaguar'].mean()
pan_mean = df['Panther'].mean()

plt.figure(figsize=(7,4))
bars = plt.bar(['Jaguar', 'Panther'],[jag_mean, pan_mean], 
               color="#570303")
plt.title('Average Resistance by Machine')
plt.ylabel('Resistance (ohms)')

#Bar Labels
plt.bar_label(bars,padding=3)
plt.margins(y=0.1)
plt.show()
```
