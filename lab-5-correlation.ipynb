```python
#imports and packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
#Load dataset
df = pd.read_csv(r'dataset\current-power.csv', 
                 delimiter=",")
df
```

```python
#Summary of statistics
df.describe()
```

```python
#Covariance of current and power
df.cov()
```

```python
#Scatter plot
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.show()
```

```python
#Correlation coefficient of current and power
df.corr()
```

```python
#Regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#Scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color="#490101",
            alpha=0.6,
            label='Data Points')

#Line plot
plt.plot(x,line,
         color="#07DCF8FF",
         label=f'Regression Line (r = {r_value:.3f})')

plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.legend()
plt.show()
```

# Air Quality

```python
#Load dataset
air = pd.read_csv(r"dataset\air-quality.csv",
                  delimiter=',')
air
```

# NOx-Humidity

```python
#Regression Line
x = air['NOx']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope * x + intercept

#Scatter plot
plt.figure(figsize=(7,4))
plt.scatter(x,y,
            color="#490101",
            alpha=0.6,
            label='Data Points')

#Line plot
plt.plot(x,line,
         color="#07DCF8FF",
         label=f'Regression Line (r = {r_value:.3f})')

plt.title('NOx vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NOx (ug/cu.m)')
plt.legend()
plt.show()
```
